Se repitieron 56  puntos, a estos se les tomo un promedio entre el primer dato y el segundo

```python
F = plt.figure(figsize=(17, 8))
graf= F.gca(projection='3d')

plot = graf.plot_trisurf(x1, y1, z1, cmap=plt.get_cmap("RdBu"))
cb = F.colorbar(plot)
graf.set_xlabel('Posicion en X')
graf.set_ylabel('Posicion en y')
graf.set_zlabel('Dato magnetometria (Nt)')
plt.show()
```

```python
F = plt.figure(figsize=(17, 8))
graf= F.gca(projection='3d')

plot = graf.plot_trisurf(x2, y2, z2, cmap=plt.get_cmap("RdBu"))
cb = F.colorbar(plot)
graf.set_xlabel('Posicion en X')
graf.set_ylabel('Posicion en y')
graf.set_zlabel('Dato magnetometria (Nt)')
plt.show()
```
